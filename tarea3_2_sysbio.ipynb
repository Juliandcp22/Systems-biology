```python
import numpy as np
import matplotlib.pyplot as plt
import random
from scipy.stats import expon
```

Fundamentos del algoritmo de Gilliespie

```python
dist_exp1= np.random.exponential(1,10000)
tau1 = np.log(1/np.random.uniform(0,1,10000))
rounded_arr1 = np.round(tau1, 1)
unique_vals1, counts1 = np.unique(rounded_arr1, return_counts=True)
```

```python
plt.hist(dist_exp1,bins= int(np.sqrt(len(dist_exp1))))
plt.title('r=1')
plt.plot(unique_vals1, counts1, label="Ï„")
plt.ylabel('Frecuencia')
plt.xlabel('Valor')
plt.legend()
plt.show()
```

```python
ta1= np.zeros(300)
tb1= np.zeros(300)
```

```python
for i in range (len(ta1)):
  ta1[i]= 1/10*np.log(1/random.uniform(0,1))
  tb1[i]= 1/5*np.log(1/random.uniform(0,1))
t= np.concatenate((ta1,tb1))
```

```python
loc1, scale1 = expon.fit(ta1)
x = np.linspace(0,5,100)
pdf_ajustada = expon.pdf(x, loc=loc1, scale=scale1)

```

```python
fig, ax = plt.subplots(1,3,figsize=(15,4))
conteos1, bins1= np.histogram(ta1, bins=int(np.sqrt(len(ta1))), density=True)
loc, scale = expon.fit(ta1)
x1 = np.linspace(bins1[0], bins1[-1], 100)
pdf_ajustada1 = expon.pdf(x1, loc=loc, scale=scale)

conteos2, bins2 = np.histogram(tb1, bins=int(np.sqrt(len(tb1))), density=True)
loc, scale = expon.fit(tb1)
x2 = np.linspace(bins2[0], bins2[-1], 100)
pdf_ajustada2 = expon.pdf(x2, loc=loc, scale=scale)


conteos3, bins3= np.histogram(t, bins=int(np.sqrt(len(t))), density=True)
loc, scale = expon.fit(t)
x3 = np.linspace(bins3[0], bins3[-1], 100)
pdf_ajustada3 = expon.pdf(x3, loc=loc, scale=scale)

ax[0].plot(x1,pdf_ajustada1,color='red')
ax[0].set_title('A')
ax[0].hist(ta1,bins=int(np.sqrt(len(ta1))),density=True)

ax[1].set_title('B')
ax[1].plot(x2,pdf_ajustada2,color='red')
ax[1].hist(tb1,bins=int(np.sqrt(len(tb1))),density=True)

ax[2].set_title('General')
ax[2].plot(x3,pdf_ajustada3,color='red')
ax[2].hist(t,bins=int(np.sqrt(len(t))),density=True)




plt.show()
```
